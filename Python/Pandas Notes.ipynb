### 30. Create own Function 

```python
# We Create Function When i don't have any logic or any method then we create function and calling by apply method in Series


# Q. Create a Column where if id%2 ==0 then Even else False
def even_odd(x):
    if x%2==0:
        return 'Even'
    return False

df['id_in_even_odd'] = df['Id'].apply(even_odd)


# Q. Create a Column where Speed in Miles from Kilometers 
def miles(x):
    return x*0.62137

df['speed_in_miles'] = df['Speed'].apply(miles)


# Q. Create a Column where Generation in int to alphabet
def int_to_alpha(y):
    if y == 1:
        return 'One'
    elif y == 2:
        return 'Two'
    elif y == 3:
        return 'Three'
    elif y == 4:
        return 'Four'
    elif y == 5:
        return 'Five'
    else:
        return 'Six'
    
df['gen_to_alphabet'] = df['Generation'].apply(int_to_alpha)

```

### 33. Pivot Table 
Pivot tables are another way of calculating grouped summary statistics. Pivot Table is similiar to Groupby Basically pivot tables are used in spreadsheets, if you have ever used spreadsheets then chances are you have used pivot tables. Let's see how to create pivot tables in pandas.

```python
walmart = pd.read_csv('walmart.csv')
walmart.drop(columns = 'Unnamed: 9', inplace = True)
```

```python
walmart.head(3)
```

```python
# Pivoting for mean Weekly_Sales for each store Type
walmart.pivot_table(index = ["Type"], values = "Weekly_Sales")       # By Default Pivot give Mean

# Pivoting for max Weekly_Sales for each store Type using pass aggfunc
walmart.pivot_table(index = ["Type"], values = "Weekly_Sales", aggfunc = max)

# Pivoting for max, min Weekly_Sales for each store Type using pass aggfunc
walmart.pivot_table(index = ["Type"], values = "Weekly_Sales", aggfunc = [max, min])

# Pivoting for mean Seekly_Sales by store Type and Holiday 
walmart.pivot_table(index = "Type", columns = 'Is_Holiday', values = "Weekly_Sales")

# Printing mean of Weekly_Sales by Department and Type
walmart.pivot_table(index = 'Department', columns = 'Type', values = 'Weekly_Sales')

# Printing mean of Weekly_Sales by Department and Type, fill missing values with 0
walmart.pivot_table(values='Weekly_Sales', index='Department', columns='Type', fill_value=0)

# Printing the of mean Weekly_Sales by Department and Type, sum all rows and cols
walmart.pivot_table(values = "Weekly_Sales", index = "Department", columns = "Type", margins = True)
```

# 36. Handling Missing Values 

To work with Missing Values you have a two way for Handle Missing Values
 - 1 -    To Drop Using Dropna()  
 
 
 - 2 -    To Fill Using Fillna()

### 36.1 Drop NAN Values Using Dropna() 

Syntax - df.dropna(axis, how, thresh, subset, inplace)
 
   - axis{0 or 'rows', 1 or ‘columns’}, By default 0 
   
    -------0, or ‘rows’ : Drop rows which contain missing values.

    -------1, or ‘columns’ : Drop columns which contain missing value.
    
   
   - how: {‘any’, ‘all’}, By default ‘any’
   
    -------if row or column is removed from DataFrame, when we have at least one NA or all NA.

    -------‘any’ : If any NA values are present, drop that row or column.

    -------‘all’ : If all values are NA, drop that row or column.
    
   
   - thresh: thresh takes integer value which tells minimum amount of na values to drop.
   
   - subset: It’s an array which limits the dropping process to passed rows/columns through list.
   
   - inplace: It is a boolean which makes the changes in data frame itself if True.
   
   
 
 
 

```python
miss = pd.read_csv('miss.csv')
miss
```

```python
miss.isnull().sum()   # No. of Null Values in Every Columns
miss.isnull().sum().sum()   # Total Null Values
```

```python
miss.dropna()             # All data deleted where NAN vlaues Exist
#or
miss.dropna(axis = 0)     # # Which rows deleted have a NAN values
```

```python
# Which Columns deleted have a NAN values
miss.dropna(axis = 1)  
```

```python
# Which Row deleted Where have a Nan Values. by default how = 'any'
miss.dropna(how = 'any')    
```

```python
# Deleted Row Where completely All NAN values Exist 
miss.dropna(how='all') 
```

```python
# We use Thresh parameter when we want a specific not null values Given a int number
# Drop NAN vlaues where at least 2 Not Null Values
miss.dropna(thresh=2) # By default axis = 0
```

```python
# single given Column name in the list where city have NAN values will be deleted
# Subset Parameter is optional
miss.dropna(subset=['city'])    
```

```python
# Multiple given Column name in the list where city and state have NAN values will be deleted
# Subset Parameter is optional
miss.dropna(subset=['city','state'])  
```

### 36.2 Fill NAN Values Using Fillna() 
syntax - fillna(value, method, axis, inplace, limit, downcast)
 - Value - Number, String, Dictionary, Series, DataFrame 
         Required, Specifies the value to replace the NULL values with.This can also be values for the entire row or column
         
 - Method - {‘backfill’, ‘bfill’, ‘pad’, ‘ffill’, None}, default None
         Method is used if user doesn’t pass any value. Pandas has different methods like bfill, backfill or ffill forward fill which fills the place with value in the Forward index or Previous/Back respectively.
         
 - Axis - {0 or ‘index’, 1 or ‘columns’}
 
        Optional, default 0. The axis to fill the NULL values along
        
 - Limit - int, default None
 
         Optional, default None. Specifies the maximum number of NULL values to fill (if method is specified)
        
 - Downcast - dict, default is None
 
        It takes a dict which specifies what dtype to downcast to which one. Like Float64 to int64.
 


```python
# All NAN values replace by 0
miss.fillna(0)

# All NAN values replace by none
miss.fillna('none')
```

```python
# Fill NAN values specific Column wise
miss.fillna({'fname': 'none', 'lname':'Null', 'age': 21}).info()
```

```python
# Fill NAN values by bfill (Backwardfill)
miss.fillna(method = 'bfill')

# Fill NAN values by ffill (Forwardfill)
miss.fillna(method = 'ffill')

# Fill NAN values by pad 
miss.fillna(method = 'pad')

# Fill NAN values by backfill
miss.fillna(method = 'backfill')
```

```python
# Fill NAN values by Axis(row wise) using method or values
miss.fillna(method = 'ffill', axis = 0)

# Fill NAN values by Axis(column wise) using method or values
miss.fillna(method = 'ffill', axis = 1)
```

```python
# Fill NAN values but specific NAN values by Limit using method or value
miss.fillna(limit = 2 ,value = 0)
```

```python
# Fill NAN value in single column
miss['age'] = miss['age'].fillna(0)
miss
```

```python
# Fill NAN value in Multiple column
miss[['age', 'city']] = miss[['age', 'city']].fillna(method = 'bfill')
miss
```
