## NumPy Exercises 

Now that we've learned about NumPy let's test your knowledge.

#### Import NumPy as np

```python
import numpy as np
```

#### Create an array of 10 zeros 

```python
np.zeros(10)
```

#### Create an array of 10 ones

```python
np.ones(10)
```

#### Create an array of 10 sevens

```python
np.ones(10)*7
```

#### Create an array of the integers from 20 to 60

```python
np.arange(20,61)
```

#### Create an array of all the even integers from 20 to 60 

```python
np.arange(20,61,2)
```

#### Create a 4x4 matrix with values ranging from 0 to 15

```python
np.arange(0,16).reshape(4,4)
```

#### Create a 4x4 identity matrix

```python
np.identity(4) #or
# np.eye(4)
```

#### Use NumPy to generate a random number between 0 and 1

```python
np.random.rand(1)
```

#### Use NumPy to generate an array of 25 random numbers sampled from a standard normal distribution

```python
np.random.randn(25)
```

#### Create the following matrix:

```python
np.linspace(0.01,1,100).reshape(10,10)
```

#### Create an array of 20 linearly spaced points between 0 and 2:

```python
np.linspace(0,2,20)
```
