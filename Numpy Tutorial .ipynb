# Numpy Functions 

```python
# ones Func

import numpy as np
arr = np.ones(5)
arr
```

```python
# zeros func

import numpy as np
arr = np.zeros(4)
arr
```

```python
# eye func

import numpy as np
arr = np.eye(5)
arr
```

```python
# diag func

import numpy as np
arr = np.diag([5,2,4,5])
arr
```

```python
# randint func

import numpy as np
arr = np.random.randint(1,10,5)
arr
```

```python
# rand func

import numpy as np
arr = np.random.rand(5)
arr
```

```python
# randn func

import numpy as np
arr = np.random.randn(5)
arr
```
